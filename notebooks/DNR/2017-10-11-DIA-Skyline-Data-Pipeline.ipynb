# Selected Reaction Monitoring Data: From RAW files to Analyses

In this notebook, I'll outline the workflow necessary to take my *Crassostrea gigas* Selected Reaction Monitoring (SRM) data from `RAW` files off of the mass spectrometer to a `.csv` usable in R Studio.

We want to ensure that the machine only captures the oyster transitions we set out to measure. ## Step 3: Evaluate Dilution Curve

The purpose of the dilution curve is to ensure the mass spectrometer methodology is specific to target proteins. Peptides that do not decrease in abundance when the concentration of protein loaded onto the machine decreases should be removed from analyses.

### Step 3a: Import Data

Under File > Import > Results, select "Add single-injection replicates in files". Navigate to the directory with RAW files. The dilution curve files are labelled "D#-O," where "D" stands for "dilution", # represents [which dilution ratio was used](https://yaaminiv.github.io/SRM-Dilution-Calculations2/), and "O" stands for "oyster." There are eight dilution files. Select **only the eight dilution curve files** and click "Open".

### Step 3b: Clean Dilution Curve Data

For each sample, go through the list of peptides and ensure Skyline selected the proper peak. Use the [predicted retetion times](https://github.com/RobertsLab/project-oyster-oa/blob/master/analyses/DNR_TransitionSelection_20170707/2017-07-08-Final-Transitions/2017-07-11-Predicted-SRM-Retention-Times.xlsx) document to identify the correct peptide peak. 

The peak selected in Skyline will have a black arrow next to it. In the example below, Skyline has selected the peak with a retention time of 27.5. It also recognizes peaks with retention times of 26.8, 27.2 and 30.7.

![image-3](https://user-images.githubusercontent.com/22335838/31088786-c094983a-a756-11e7-84b5-8e8a81ae48a4.png)

If the correct peak is selected, adjust the peak boundaries (dashed lines) so they only surround the peak. This can be done simply by dragging the boundary to the desired ocation.

If the incorrect peak is selected, you will need to manually select the correct peak. If Skyline recognizes the peak, simply click on the correct peak. If Skyline does not recognize the peak, hover over the x-axis (Retentiom time). The mouse will turn into a crosshair. Select the proper peak by dragging the crosshair over the retention times that the peak spans.

If there is no discernable peak present (the transitions do not elute at the same time), right click and select "Remove peak."

After selecting the correct peak, right click and select "Apply Peak to All." This may not correctly identify the proper peak for all samples, but for a good majority. You will still need to check all peptide peaks and adjust peak boundaries.

### Step 3c: Identify excludable peptides

Once the proper peaks are selected, examine peak areas across the dilution curve. D1-O has the highest concentration of oyster protein, and D8-O has the lowest concentration. As the concentration of protein loaded decreases, so should peak area:

![30131820-0838676a-9303-11e7-9068-f5f26c802bd0](https://user-images.githubusercontent.com/22335838/30982905-abf5a318-a43d-11e7-97a1-aa166e0a6e67.png)

If this is not the case, then the peptide needs to be exluced from analyses.

The final dilution curve document can be found [here](http://owl.fish.washington.edu/spartina/DNR_SRM_20170728/Analyses/2017-09-06-GigasDilutionCurve.sky.zip). Transitions kept after dilution curve analyses should match those found in [this transition list](https://github.com/RobertsLab/project-oyster-oa/blob/master/analyses/DNR_SRM_20170902/2017-10-11-Final-SRM-Transition-List-with-PRTC.csv).

## Step 4:  Create SRM Skyline Document

This is the Skyline document used to view mass spectrometry data from oyster samples. This is also the primary document used for viewing sample data and assessing peak quality.

### Step 4a: Duplicate the Dilution Curve Document

Create a copy of the dilution curve Skyline document and rename it for SRM data.

### Step 4b: Delete Peptides Based on Dilution Curve

If there were any peptides that did not follow the expected pattern in the dilution curve, remove them from the analyte tree. To do this, right click on the peptide, then select "Delete."

Ensure all target proteins have at least two associated peptides before proceeding. If not, delete the protein from analysis as it no longer meets the criteria of a target.

### Step 4c: Add SRM Raw Data

Under Edit >> Manage Results, remove all RAW files associated with the dilution curve. Under File > Import > Results, select "Add single-injection replicates in files". Navigate to the directory with RAW files. The files associated with samples are named "2017_July_10_bivalves_NUMBER." 

Import files numbered 2 through 137. Select files to import and click "Open".

## Step 5: Clean Data in Skyline

In this step, files without any data will be removed and peaks in Skyline will be verfied against predicted retention times.

### Step 5a: Remove Blanks

Blank solutions without any oyster protein were run between sets of samples and should be excluded from analysis. Under Edit >> Manage Results, delete these files from the Skyline document. Blanks can be found using the [sample key](http://owl.fish.washington.edu/spartina/DNR_SRM_20170728/Analyses/2017-07-28-SRM-Sequence-File.xlsx). In the "Comment" column, "blank" indicates a file contains no oyster protein.

### Step 5b: Remove Samples Without Any Data

There could be sample files that did not pick up any peptide data whatsoever, even though they were not blanks. Thumb through the different sample results. If any samples display the phrase "No product ion chromatograms found," for all peptides, remove the sample file from the document.

Using the [sample key](http://owl.fish.washington.edu/spartina/DNR_SRM_20170728/Analyses/2017-07-28-SRM-Sequence-File.xlsx), identify that file's technical replicate and delete is as well. Technical replicates will have the same oyster ID in the "Comment" column (ex. O01-1 and O01-2).

Similar to the dilution curve, all peptide peaks need to be verified for each sample. See Step 3a for instructions.

## Step 7: Analyze Technical Replication

Each oyster sample was run in technical duplicates. Oyster protein sasmples [prepared](https://github.com/RobertsLab/resources/blob/master/protocols/ProteinprepforMSMS.md) prior to mass spectrometry were placed in individual autosampler vials. Each vial was injected twice, producing two technical replicates. Before proceeding, it is important to verify that technical replicates have the same peak area for all peptides and transitions. If this is not the case, those samples must be removed from the analysis.
